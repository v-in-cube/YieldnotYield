```python
import matplotlib.pyplot as plt
import matplotlib.animation
from matplotlib.animation import FuncAnimation
import glob
```

```python
# Define a function to update the frame
def update_frame(frame):
    # Load the PNG file for the current frame
    file_path = file_list[frame]
    img = plt.imread(file_path)
    # Clear the axis and plot the new image
    ax.cla()
    ax.imshow(img)
    ax.set_aspect('equal')  # Set the aspect ratio to preserve image proportions
    ax.axis('off')  # Hide the axis labels and tick marks
```

```python
# Define a function to extract the frame number from the filename
def extract_frame_number(filename):
    return int(filename.split('_')[-2])
```

```python
# Create a figure and axis
fig, ax = plt.subplots()
fig.patch.set_facecolor('white')
ax.set_facecolor('white')
# Get the list of files in the desired order
file_list = sorted(glob.glob('hte_hte_*_epoch.png'), key=extract_frame_number)

# Determine the number of frames
num_frames = len(file_list)
frames=range(num_frames)
# Create the animation
anim = FuncAnimation(fig, update_frame, frames, interval=100)

from IPython.display import HTML
plt.close(anim._fig)
HTML(anim.to_jshtml())
```
